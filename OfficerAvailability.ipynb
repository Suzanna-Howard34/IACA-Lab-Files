```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import CAD2018.xlsx
df = pd.read_excel('CAD2018.xlsx')
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.info()
```

```python
# sort by Unit then Date/Time
df = df.sort_values(by=['Unit', 'Date/Time'])
df.head()
```

```python
# create a new column called Time that subtracts the current Date/Time from the previous Date/Time
df['Time'] = df.groupby('Unit')['Date/Time'].diff()
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
# remove any rows that have a value in the Time column over 5:00:00
df = df[df['Time'] <= pd.Timedelta(hours=5, minutes=0, seconds=0)]
df.head()
```

```python
df.shape
```

```python
# create a new column called Date that includes only the date from the Date/Time column
df['Date'] = df['Date/Time'].dt.date
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# convert Date column to a Date format
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df.info()
```

```python
# Create a new column to store the previous status for each unit
df['Previous_Status'] = df.groupby('Unit')['Status'].shift(1)

# Filter the DataFrame
filtered_df = df[(df['Status'] == 'ENRT') & (df['Previous_Status'] == 'CMPLT')]

# Display the filtered DataFrame
display(filtered_df.head())
```

```python
df.shape
```

```python
filtered_df.shape
```

```python
# rename filtered_df to dff
dff = filtered_df
```

```python
dff.shape
```

```python
# Create a new column called SUM, which sums the Time column, group by Unit and Date
dff['SUM'] = dff.groupby(['Unit', 'Date'])['Time'].transform('sum')

# Create a new column called AVG, which averages the Time column, group by Unit and Date
dff['AVG'] = dff.groupby(['Unit', 'Date'])['Time'].transform('mean')

# Create a new column called SumAVG, which averages the SUM column, group by Unit
dff['SumAVG'] = dff.groupby('Unit')['SUM'].transform('mean')

# Create a new column called AVGAvail, which averages the AVG column, group by Unit
dff['AVGAvail'] = dff.groupby('Unit')['AVG'].transform('mean')


display(dff.head())
```

```python
dff.shape
```

```python
# remove duplicate Units
dff = dff.drop_duplicates(subset=['Unit'])
dff.shape
```

```python
display(dff.head(16))
```

```python
# run descriptive statistics on AVGAvail
dff['AVGAvail'].describe()
```

```python
# export DFF to Excel
dff.to_excel('OfficerAvail2018.xlsx', index=False)

# export descriptive statistics of AVGAvail to Excel
dff['AVGAvail'].describe().to_excel('Descriptive_Statistics2018.xlsx')
```
