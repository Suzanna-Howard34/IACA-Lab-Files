```python
import pandas as pd
import glob
import os

# If files are in your Google Drive, mount it first:
# from google.colab import drive
# drive.mount('/content/drive')

# Path to your Excel files (adjust if using Google Drive)
path = "/content/"   # folder where your 12 Excel files are stored
all_files = glob.glob(os.path.join(path, "*.xlsx"))

dfs = []

for file in all_files:
    # Read Excel file
    df = pd.read_excel(file)

    # Extract filename without path & extension
    zone_name = os.path.splitext(os.path.basename(file))[0]

    # Add Zone column
    df["Zone"] = zone_name

    dfs.append(df)

# Merge all into one DataFrame
merged_df = pd.concat(dfs, ignore_index=True)

# Save to a new Excel file
merged_df.to_excel("merged_with_zone.xlsx", index=False)
```

```python
# convert Report Date/Time and Occurred On to Date/Time format
df["Report Date/Time"] = pd.to_datetime(df["Report Date/Time"])
df["Occurred On"] = pd.to_datetime(df["Occurred On"])
```

```python
# insert a column called DOW for Day of the Week name of Report Date/Time
df.insert(2, "DOW", df["Report Date/Time"].dt.day_name())

# insert a column called Hour for the hour of day of Report Date/Time
df.insert(3, "Hour", df["Report Date/Time"].dt.hour)

df.head()
```

```python
df.shape
```

```python
# sort by Zone then Report Date/Time ascending
df = df.sort_values(by=["Zone", "Report Date/Time"])
df.head()
```

```python
df.tail()
```

```python
# Insert a new column called DBH which calculates the date difference from the line above it, group by Zone
df["DBH"] = df.groupby("Zone")["Report Date/Time"].diff().dt.days
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Calculate difference in days (with fractions) and round to 2 decimals
df["DBH"] = (
    df.groupby("Zone")["Report Date/Time"]
      .diff()
      .dt.total_seconds()
      .div(86400)  # 86400 seconds in a day
      .round(2)    # round to 2 decimal places
)
df.head()
```

```python
# prompt: descriptive statistics on DBH by Zone
# Group data by 'Zone' and get descriptive statistics for 'DBH'
zone_dbh_stats = df.groupby('Zone')['DBH'].describe()
zone_dbh_stats
```

```python
import pandas as pd
from datetime import datetime, timedelta

# --- Example: you already have zone_dbh_stats ---
# zone_dbh_stats = df.groupby("Zone")["DBH"].describe()

# Calculate last Report Date/Time for each Zone
last_dates = df.groupby("Zone")["Report Date/Time"].max().reset_index()
last_dates.rename(columns={"Report Date/Time": "Last Crime"}, inplace=True)

# Merge into your stats
zone_dbh_stats = zone_dbh_stats.merge(last_dates, on="Zone")

# Add Mean+STD
zone_dbh_stats["Mean+STD"] = zone_dbh_stats["mean"] + zone_dbh_stats["std"]

# Add Forecast = Last Crime + (Mean+STD days)
zone_dbh_stats["Forecast"] = zone_dbh_stats["Last Crime"] + pd.to_timedelta(
    zone_dbh_stats["Mean+STD"], unit="D"
)

# Add Todays date (when report is run)
zone_dbh_stats["Today"] = pd.to_datetime(datetime.today().date())

# Add Due = (Forecast - Today) in days
zone_dbh_stats["Due"] = (
    (zone_dbh_stats["Forecast"] - zone_dbh_stats["Today"]).dt.days
)

zone_dbh_stats.head()
```

```python
# export 'zone_dbh_stats' to Excel
zone_dbh_stats.to_excel("zone_dbh_stats.xlsx", index=False)
```
