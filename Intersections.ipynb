## Extract and sort parts

### Subtask:
For each identified location, extract the two number-direction parts, convert the numbers to integers, sort the parts based on the numerical value, and then rejoin them with a space.

```python
import re

def sort_location_numerical(location):
    # Extract number-direction parts
    parts = re.findall(r'\d+\s*[WSsw]', location)
    if len(parts) == 2:
        # Extract number and direction, and convert number to int
        parsed_parts = []
        for part in parts:
            match = re.match(r'(\d+)\s*([WSsw])', part)
            if match:
                number, direction = match.groups()
                parsed_parts.append((int(number), direction.upper(), part))
        # Sort based on the numerical value
        sorted_parts = sorted(parsed_parts)
        # Rejoin the original parts in sorted order
        return f"{sorted_parts[0][2]} {sorted_parts[1][2]}".strip()
    return location

df_specific_intersections['Sorted Location'] = df_specific_intersections['Location'].apply(sort_location_numerical)
display(df_specific_intersections.head())
```
